```python
def ip_to_ints(df, column):
    l = [x.strip(" ").split(".") for x in df[column].unique()]
    for x in l:
        assert len(x) == 4
    l = list(map(list, zip(*l)))
    d = {}
    for i in range(4):
        for x in l[i]:
            if x not in d:
                d[x] = len(d)
    l0, l1, l2, l3 = [], [], [], []
    for x in df[column]:
        x = x.strip(" ").split(".")
        l0.append(d[x[0]])
        l1.append(d[x[1]])
        l2.append(d[x[2]])
        l3.append(d[x[3]])
    df[column + '0'] = pd.Series(l0, index = df.index)
    df[column + '1'] = pd.Series(l1, index = df.index)
    df[column + '2'] = pd.Series(l2, index = df.index)
    df[column + '3'] = pd.Series(l3, index = df.index)
    return df
```

```python
ip_list = ['ip', 'srcip', 'dstip']
for column in ip_list:
    df_merged = ip_to_ints(df_merged, column)
columns_to_drop += ip_list
```

```python
def str_to_ints(df, column):
    print(column)
    l = [str(x).strip(" ") for x in df[column].unique()]
    d = {}
    for x in l:
        d[x] = len(d)
    l = []
    for x in df[column]:
        x = str(x).strip(" ")
        l.append(d[x])
    df[column + '_int'] = pd.Series(l, index = df.index)
    return df
```

Subset data based on max seq length

```python
# write merged data to drive
# since size is huge, only using 25% of data
def count_y(df):
    y = df['notified'].values
    print("Counts of label '1': {}".format(sum(y == 1))) 
    print("Counts of label '0': {} \n".format(sum(y == 0))) 

df_merged = pd.read_csv(path + 'data/merged_full_my.csv', header=[0], dtype=np.int32)
alert_id_dict = {}
for i, row in df_merged.iterrows():
    alert_id = row['alert_ids_int']
    if alert_id in alert_id_dict:
        alert_id_dict[alert_id].append(i)
    else:
        alert_id_dict[alert_id] = [i]

MAX_SEQ_LEN = 1000

cnt1 = 0
cnt2 = 0
sum = 0
for alert_id in alert_id_dict:
    # alert_id_dict[alert_id] = get_subset(alert_id_dict[alert_id])
    if len(alert_id_dict[alert_id]) > MAX_SEQ_LEN:
        cnt1 += 1
    else:
        cnt2 += 1
    sum += min(len(alert_id_dict[alert_id]), MAX_SEQ_LEN)
print(cnt1, cnt2, sum)
```

```python
def get_subset(l):
    size = min(len(l), MAX_SEQ_LEN)
    x = np.random.choice(len(l), size=size, replace=False)
    l = np.array(l)
    l = sorted(l[x])
    return list(l)

np.random.seed(0)
subset_list = []
for alert_id in alert_id_dict:
    subset_list += get_subset(alert_id_dict[alert_id])

print(df_merged.shape)
df_merged = df_merged.iloc[subset_list]
print(df_merged.shape)
```

Dataset 2 -- Oversampling using SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
def read_file(file_path):
    return pd.read_csv(file_path, header=[0], dtype=np.int32)

df1 = read_file(path + 'data/dataset_1/train_data.csv')
column_names = list(df1.columns)
df2 = read_file(path + 'data/dataset_1/val_data.csv')
frames = [df1, df2]
x = pd.concat(frames).values

df1 = read_file(path + 'data/dataset_1/train_label.csv')
df2 = read_file(path + 'data/dataset_1/val_label.csv')
frames = [df1, df2]
y = pd.concat(frames).values
```

```python
print("Before OverSampling, counts of label '1': {}".format(sum(y == 1))) 
print("Before OverSampling, counts of label '0': {} \n".format(sum(y == 0))) 
  
sm = SMOTE(random_state = 0) 
x_new, y_new = sm.fit_sample(x, y.ravel()) 

print('After OverSampling, the shape of train_x: {}'.format(x_new.shape)) 
print('After OverSampling, the shape of train_y: {} \n'.format(y_new.shape)) 
  
print("After OverSampling, counts of label '1': {}".format(sum(y_new == 1))) 
print("After OverSampling, counts of label '0': {}".format(sum(y_new == 0)))
```

```python
df = pd.DataFrame(data=x_new, columns=column_names)
df['notified'] = list(y_new)
l = list(df.columns)
for i in range(len(l)):
    print(i, l[i])
df.head()
```

```python
df.sort_values(by = [df.columns.values[72], df.columns.values[54]], ascending=[True, True], inplace=True)
```

```python
train, val = train_test_split(df, test_size=0.176470588, shuffle = False)

df_to_file(train, path + 'data/dataset_2/train')
df_to_file(val, path + 'data/dataset_2/val')
```
